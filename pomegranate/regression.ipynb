```python
clf = linear_model.LinearRegression()
```

```python
observation = []
price = []
for index, row in new_resale.iterrows():
    observation.append([int(row["FLOOR AREA (SQM)    "])])
    re_price  = row["RESALE PRICE   "].replace("$","").replace(",","").replace(".00","")
    price.append([float(re_price)])

np_observation = np.array(observation)
np_price = np.array(price)
```

```python
clf.fit(np_observation, np_price)
```

```python
from sklearn.preprocessing import Imputer
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp = imp.fit(np_observation)

# Impute our data, then train
np_observation_imp = imp.transform(np_observation)
```
